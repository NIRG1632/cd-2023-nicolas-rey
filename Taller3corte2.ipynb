#### Regresión Logística

Hallando el Trio entre las columnas Age, Sex, SibSp, Pclass, Embarked que tenga el mayor Pseudo R-squ

```python
aux = set()
max = 0
for i in df.iloc[:,[2,5,6,11,12]]:
  aux2 = set()
  aux.add(i)
  for j in df.iloc[:,[2,5,6,11,12]]:
    aux2.add(j)
    for k in df.iloc[:,[2,5,6,11,12]]:
       if (i != j and j != k and k != i and not(k in aux2) and not(j in aux)):
         
        #Logit Regression
        x = df.loc[:,[i,j,k]]
        y = df.Survived
        logit_model = sm.Logit(y,x)
        result = logit_model.fit()
        print(result.summary())
        
        if max < result.prsquared:
          max = result.prsquared
          columnas = i + ' ' + j + ' ' + k
print('El mejor Pseudo R-squ es {} vs Survived: {}'.format(columnas,max))
```

RTA: El mejor Trio es Pclass, SibSp y Sex2 y se obtiene un Pseudo R-squ de 0.29389178324802556

```python
from sklearn.model_selection import train_test_split

#Predicciones con scikit-learn
train, test = train_test_split(df, test_size=0.20)
print('Entreno: ', len(train))
print('Prueba: ', len(test))
```

```python
df.info()
```

## Usando las 3 columnas Pclass, SibSp, Sex2

```python
x_train = train.iloc[:,[2, 6, 11]]
y_train = train.Survived
x_test = test.iloc[:,[2, 6, 11]]
y_test = test.Survived
```

```python
logit_model=sm.Logit(y_train, x_train)
result=logit_model.fit()
print(result.summary())
```

```python
yhat = result.predict(x_test)
prediccion = list(map(round, yhat))

print('Reales:     ', list(y_test))
print('Prediccion: ', prediccion)
```

```python
true_positive = 0
false_negative = 0
true_negative = 0
false_positive = 0
for predi, real in zip(prediccion, list(y_test)):
  # Verdadero positivo sobrevivió, sobrevivió
  if predi == 1 and real == 1:
    true_positive += 1
  # Falso positivo sobrevivió, murió
  if predi == 1 and real == 0:
    false_positive += 1
  # Verdadero negativo murió, murió
  if predi == 0 and real == 0:
    true_negative += 1
  # Falso negativo murió, sobrevivió
  if predi == 0 and real == 1:
    false_negative += 1

print('Verdadero Positivo: ', true_positive)
print('Falso Positivo: ', false_positive)
print('Verdadero negativo', true_negative)
print('Falso negativo', false_negative)
```

```python
precision = true_positive/(true_positive + false_positive)
recall = true_positive/(false_negative + true_positive)

print('Precisión: ', precision)
print('Recall', recall)
```
